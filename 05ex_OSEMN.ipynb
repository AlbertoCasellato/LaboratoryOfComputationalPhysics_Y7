1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
random_numbers = npr.random(10)
with open("simple_data.txt", "w") as f:
    for number in random_numbers:
        f.write(f"{number}\n")
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
random_matrix = npr.random((10,5))
with open("data.txt", "w") as f:
    for row in random_matrix:
        f.write(f"{row}\n")
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
new_data = ""
with open("data.txt", "r") as f:
    for line in f.readlines():
        row = [val for val in line[1:-2].split(" ") if val != ""]
        for val in row:
            new_data += val + ","
        new_data += "\n"

with open("data.csv", "w") as f:
    f.write(new_data)
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
cards = []
for n in np.genfromtxt("credit_card.dat", dtype=str)[:-1]:
    card_n = int("".join([c for c in [chr(int(n[i:i+6], 2)) for i in range(0, len(n), 6)] if c not in [" ", "\n"]]))
    cards.append(card_n)
cards
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
filtered_data = []
for data in json.load(open("user_data.json")):
    if data["CreditCardType"] == "American Express":
        filtered_data.append(data)


with open("user_data_filtered.csv", "w") as f:
    header = ",".join(list(filtered_data[0].keys())) + "\n"
    f.write(header)
    for data in filtered_data:
        d = ",".join(list(data.values())) + "\n"
        f.write(d)
```
